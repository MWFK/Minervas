### Text Answer

```python
def getAnswer(instruction, question):

  start_time = time.time()

  completion = client.chat.completions.create(
    model="gpt-4",
    messages=[
      {"role": "system", "content": instruction},
      {"role": "user", "content": question}],
    max_tokens=10,
  )

  response_time = time.time() - start_time

  return completion, response_time
```

```python
# Test text

instruction = "You are an artificial intelligence and data strategy expert, skilled in explaining complex programming and technical concepts with precise flair."
question    = "how to build an AI Data strategy using microsoft azure"

completion, response_time = getAnswer(instruction, question)
answer = completion.choices[0].message.content

print(f"This is your answer: {answer}")
print(f"\nFull response received {response_time:.2f} seconds after request")
```

### Audio Answer

```python
def getAudio(answer):

    start_time = time.time()

    response = client.audio.speech.create(
    model="tts-1",
    voice="echo",
    input=answer
    )

    response_time = time.time() - start_time

    return response, response_time
```

```python
### Test audio

audio_response, audio_response_time = getAudio(answer)

speech_file_path = Path('content').parent / 'audio.mp3'
audio_response.stream_to_file(speech_file_path)

print(f"\nFull response received {audio_response_time:.2f} seconds after request")
audio = Audio(speech_file_path)
audio
```
